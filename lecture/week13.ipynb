# 2. Matplotlib
- cf. 일반적으로 시각화할 땐 데이터프레임 자료형 활용

```python
# 시각화 옵션
# 제목
plt.plot(xdata, ydata)
plt.title("Line Graph")
plt.show()
```

```python
# 범례 (2개 이상의 데이터를 표현할 때)
plt.plot(xdata, label = 'asc1')
plt.plot(ydata, label = 'asc2')
plt.title("Line Graph")
plt.legend() # 라벨의 범례 표시
plt.show()
```

```python
# 색상
plt.plot(xdata, ydata, color = 'red')
plt.title("Line Graph")
plt.show()
```

```python
# x축 및 y축 이름
plt.plot(xdata, ydata, color = 'red')
plt.title("Line Graph")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```

```python
# 그래프 선 모양
plt.plot(xdata, label = 'asc1', color = 'r', linestyle = '--')
plt.plot(ydata, label = 'asc2', color = 'g', linestyle = ':')
plt.title("Line Graph")
plt.legend() # 라벨의 범례 표시
plt.show()
```

```python
xdata = [10, 20, 30, 40]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color = 'b', linestyle = '--', 
         marker = 'o', markersize = '10')
plt.show()
```

```python
# 그림 범위 지정 (축 범위)
xdata = [10, 20, 30, 40]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color = 'b', linestyle = '--', 
         marker = 'o', markersize = '10')

plt.xlim(0, 50)
plt.ylim(-5, 15)
plt.show()
```

```python
import os
data = pd.read_excel(os.getcwd() + '/data2/fine_dust.xlsx', 
                   index_col='area')
data
```

```python
data2020 = data[2020]
data2020
```

```python
# 선 그래프: 시간에 따라 연속적으로 변하는 데이터를 표현하기에 가장 적합
plt.figure(figsize=(15, 10)) # 가로, 세로 크기 지정
plt.plot(data2020, color= 'b', marker= 'o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2020 Fine Dust Line Graph')
plt.grid() # 배경 격자
plt.show()
```

```python
plt.figure(figsize=(15, 10)) # 가로, 세로 크기 지정

for year in range(2012, 2021):
    chartdata = data[year]
    plt.plot(chartdata,  marker= 'o', label=year)

plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2020 Fine Dust Line Graph')
plt.legend()
plt.grid() # 배경 격자
plt.show()
```

```python
# 막대 그래프: 집단별 차이 (크고 작음의 차이를 한 눈에 파악)
## 항목이 적으면 가로, 많으면 세로가 적합

# 가로 막대그래프
plt.figure(figsize=(15, 5))
plt.bar(data2020.index, data2020, color='g')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2020 Fine Dust Bar Graph')
plt.ylim(25, 40) # 값 차이를 좀 더 한 눈에 볼 수 있게 y축 범위 조정
plt.grid()
plt.show()
```

```python
# 산점도 그래프: 두 변수간의 상관관계를 대략적으로 파악
data = pd.read_excel(os.getcwd() + '/data2/heatlh_screenings_2020.xlsx')
```

```python
data
```

```python
hdata = data['height']
wdata = data['weight']
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(hdata, wdata)

plt.xlabel('height')
plt.ylabel('weight')
plt.title('2020 Health Screenings Scatter Graph')
plt.grid()
plt.show()
```

```python
hd1_data = data['HDL']
ldl_data = data['LDL']
chole_data = data['cholesterol']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(hd1_data, ldl_data, color='r', edgecolors= 'w', label='HDL*LDL')
plt.scatter(hd1_data, chole_data, color='g', edgecolors= 'w', label='HDL*Chole')
plt.scatter(chole_data, ldl_data, color='b', edgecolors= 'w', label='Chole*LDL')

plt.xlabel('HDL')
plt.ylabel('LDL')
plt.title('2020 Health Screenings Scatter Graph')
plt.grid()
plt.legend()
plt.show()
```

```python
# 히스토그램: 변수가 하나인 데이터의 빈도 분석 

mdata = data.loc[data.gender == 1, ['gender', 'height']]
mdata
```

```python
plt.figure(figsize=(10,6))
plt.hist(mdata['height'], label='Male') # cf. 구간 크기 bins

plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 Health Screening Male Height Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
fdata = data.loc[data.gender == 2, ['gender', 'height']]
fdata
```

```python
plt.figure(figsize=(10,6))
plt.hist(mdata['height'], label='Male') # cf. 구간 크기 bins
plt.hist(fdata['height'], label='Female')

plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 Health Screening Male Height Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
# 누적 히스토그램
x = np.array([mdata['height'], fdata['height']])
```

```python
plt.figure(figsize=(10,6))
plt.hist(x, label= ['Male', 'Female'], stacked=True)

plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 Health Screening Male Height Stacked Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
# 상자그림 그래프: 극단치 확인할 수 있는 그래프
mdata = np.array(mdata['height'])
fdata = np.array(fdata['height'])

plt.figure(figsize = (8, 5))
plt.boxplot([mdata, fdata], labels= ['Male', 'Female'])

plt.xlabel('gender')
plt.ylabel('height')
plt.title('2020 Health Screening Male & Female Height Box Plot')
plt.show()
```

```python
# 가로로 상자그림 그래프
# 상자그림 그래프: 극단치 확인할 수 있는 그래프
plt.figure(figsize = (8, 5))
plt.boxplot([mdata, fdata], labels= ['Male', 'Female'], vert=False)

plt.xlabel('gender')
plt.ylabel('height')
plt.title('2020 Health Screening Male & Female Height Box Plot')
plt.show()
```
